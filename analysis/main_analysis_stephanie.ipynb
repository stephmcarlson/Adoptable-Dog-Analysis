```python
access_token = token['access_token']

headers = {
    "Authorization": f"Bearer {access_token}"
}

base_url = "https://api.petfinder.com/v2/animals"

dog_response_list = []

params = {
"type":"dog",
"location" : "Chicago, Illinois"
}

dog_response = requests.get(base_url, headers=headers, params=params)
dog_response.json()


```

```python
# Check Pagination
r = dog_response.json()
r['pagination']

# Create Empty Data Frame to Store Data
df_all_dogs = pd.DataFrame()
pages = r['pagination']['total_pages']

# Loop through all possible pages in response
for i in range(1,pages):
    params['page'] = i
    try:
        response = requests.get(base_url, headers=headers,params=params)
        dframe = pd.DataFrame(response.json()['animals'])
        df_all_dogs = pd.concat([df_all_dogs,dframe],axis=0)
    except Exception as e:
        print(e)
```

```python
# Save Raw Dataset
df_all_dogs.to_csv('raw_data.csv',index=False)
```
